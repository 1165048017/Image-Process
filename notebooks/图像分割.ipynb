```python

import cv2
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
import os
from imp import reload

import seaborn as sns
sns.set_style("dark")

%matplotlib inline

img_base_path = r"../image_data/DIP3E_Original_Images_CH10"
```

## Marr-Hildreth边缘检测

### 原图

```python
img_path = os.path.join(img_base_path,r"Fig1016(a)(building_original).tif")

img = Image.open(img_path)
img = np.asarray(img)
print(img.shape)

plt.imshow(img,cmap="gray")
```

### 高斯模糊

```python
tmp_img = cv2.GaussianBlur(img,(9,9),1.5)
plt.imshow(tmp_img,cmap = "gray")
```

### 拉普拉斯变换

```python
plt.imshow(cv2.Laplacian(tmp_img,ddepth = -1,ksize = 3))
plt.show()

plt.imshow(cv2.Laplacian(img,ddepth = -1,ksize = 3))
plt.show()
```
